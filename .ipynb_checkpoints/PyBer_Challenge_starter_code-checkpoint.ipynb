### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head()
city_data_df.head()
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
ride_count
```

```python
# 2. Get the total drivers for each city type
driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
driver_count
```

```python
#  3. Get the total amount of fares for each city type
fare_amount = pyber_data_df.groupby(["type"]).sum()["fare"]
fare_amount
```

```python
# Get the average fare per ride by city type. 
fare_average = pyber_data_df.groupby(["type"]).mean()["fare"]
fare_average
```

```python
# 5. Get the average fare per driver for each city type. 
driver_average = pyber_data_df.groupby(['type']).sum()['fare']/city_data_df.groupby(['type']).sum()['driver_count']
driver_average
```

```python
# 6. Creating a data frame from this info

pyber_summary = { 'Total Rides': ride_count, 'Total Drivers': driver_count, 
                 'Total Fares': fare_amount, 'Average Fare per Ride': fare_average,
                'Average Fare per Driver': driver_average}
pyber_summary_types_df = pd.DataFrame(pyber_summary)
pd.options.display.float_format = '${:,.2f}'.format
pyber_summary_types_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_types_df.index.name = None
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fare_date_amount = pyber_data_df.groupby(["type",'date']).sum()["fare"]


fare_date_amount.head()
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_data_df = pyber_data_df.reset_index()
pyber_data_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_data_df_pivot = pyber_data_df.pivot(index = 'date', columns = 'type', values = 'fare')
pyber_data_df_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_pivot_df = pyber_data_df_pivot.loc["2019-01-01":'2019-04-29']
pyber_pivot_df.head(8)
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_pivot_df.index = pd.to_datetime(pyber_pivot_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_pivot_week_df = pyber_pivot_df.resample('W').agg([sum])
pyber_pivot_week_df.head()
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

#get city type fare amount by week. 
urban_fare_week = pyber_pivot_week_df.groupby(["date"]).sum()["Urban"]
suburban_fare_week = pyber_pivot_week_df.groupby(["date"]).sum()["Suburban"]
rural_fare_week = pyber_pivot_week_df.groupby(['date']).sum()['Rural']

# Create a plot for the urban cities fare by week.
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(urban_fare_week, color="cyan", label = "Urban")
plt.legend(loc='best')
# Add the title, y-axis label and grid.
ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare ($USD)')
ax.set_xticks(['2019-01-01','2019-02-01','2019-03-01','2019-04-01'])
ax.set_xticklabels(['Jan','Feb','Mar','Apr'])


#Create the suburban fare amount graph
ax.plot(suburban_fare_week, color="teal", label = 'Suburban')
plt.legend(loc='best')
# Add the title, y-axis label and grid.
ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare ($USD)')
ax.set_xticks(['2019-01-01','2019-02-01','2019-03-01','2019-04-01'])
ax.set_xticklabels(['Jan','Feb','Mar','Apr'])


# Create a plot for the rural cities fare by week. 
ax.plot(rural_fare_week, color="lightskyblue", label = 'Rural')
plt.legend(loc='best')
# Add the title, y-axis label and grid.
ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare ($USD)')
ax.set_xticks(['2019-01-01','2019-02-01','2019-03-01','2019-04-01'])
ax.set_xticklabels(['Jan','Feb','Mar','Apr'])


# Save the figure.
plt.savefig("analysis/Pyber_fare_summary.png")


plt.show()
```
